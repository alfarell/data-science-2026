# Hands On Pertemuan 2 - pd.read_csv()

## Load dan Inspect Dataset

```python
import pandas as pd

# Memuat dari URL (dataset publik)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

# Q1 - Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?

```python
print("Shape (baris, kolom): ", df.shape) # mengambil jumlah baris dan kolom pada data csv yang dimuat
print("List Kolom: ", df.columns.tolist()) # mengambil daftar kolom yang terdapat pada data csv yang dimuat
```

# Q2 - Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?

```python
missing = df.isnull().sum() # mengambil kolom mana saja yang memiliki field yang kosong diantara semua baris data
percent = (df.isnull().sum() / len(df)) * 100 # menjumlahkan lalu membuat rata-rata dari total data yang kosong dengan jumlah keseluruhan data

missing_df = pd.DataFrame({
    'Jumlah field kosong': missing,
    'Persentase (%)': percent.round(2)
})

print(missing_df[missing_df['Jumlah field kosong'] > 0])
```

# Q3 - Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas.

```python
survived_pct = df['Survived'].mean() * 100 # mengambil persentase dari semua penumpang yang selamat
print(f"Persentase penumpang yang selamat: {survived_pct.round(2)}%")
```

# Q4 - Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?

```python
female_c1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)] # mengambil data dengan gender "female" dan pclass "1"
total_female_c1 = len(female_c1) # mengambil jumlah semua data dari data di atas
avg_age_female_c1 = female_c1['Age'].mean() # mengambil rata-rata usia dari keseluruhan data di atas

print(f"Analisis Wanita Kelas 1")
print(f"- Jumlah Penumpang: {total_female_c1}")
print(f"- Rata-rata Usia  : {avg_age_female_c1.round()} tahun")
```

# Q5 - Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?

```python
print("Tingkat keselamatan per kelas penumpang")
survived_by_class = df.groupby('Pclass')['Survived'].mean() * 100 # mengambil rata-rata penumpang selamat per kelas penumpang
print(survived_by_class)
print(f"Kelas dengan tingkat keselamatan tertinggi yaitu: Kelas {survived_by_class.idxmax()}")
```

# Q6 - Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik. Sertakan interpretasi singkat.

```python
result = df.groupby('Sex').agg(
  jumlah = ('PassengerId', 'count'),
  selamat = ('Survived', 'sum'),
  rata_usia = ('Age', 'mean'),
)
result = result.round()
print(result)
print("Dari data tersebut, jumlah penumpang wanita lebih sedikit daripada jumlah penumpang laki-laki, namun penumpang wanita memiliki jumlah korban selamat lebih banyak daripada laki-laki.")
```
